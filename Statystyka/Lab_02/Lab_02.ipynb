```python
import numpy as np
```

```python
wartosci = np.array([1, 2, 3, 4, 5, 6])
prawdopodobienstwo = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
```

```python
srednia = np.sum(wartosci * prawdopodobienstwo)
wariancja = np.sum(prawdopodobienstwo * (wartosci - srednia)**2)
odchylenie_standardowe = np.sqrt(wariancja)
```

```python
print(f'Średnia: {srednia}')
print(f'Wariancja: {wariancja}')
print(f'Odchylenie standardowe: {odchylenie_standardowe}')
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

p_bernoulli = 0.3
n_dwumianowy = 10
p_dwumianowy = 0.4
lambda_poisson = 3
```

```python
proby_bernoulli = np.random.choice([0, 1], size=100, p=[1-p_bernoulli, p_bernoulli])
proby_dwumianowy = np.random.binomial(n=n_dwumianowy, p=p_dwumianowy, size=100)
proby_poisson = np.random.poisson(lambda_poisson, size=100)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(proby_bernoulli, bins=[-0.5, 0.5, 1.5], edgecolor='black', alpha=0.7)
plt.title('Rozkład Bernoulliego')

plt.subplot(1, 3, 2)
plt.hist(proby_dwumianowy, bins=np.arange(-0.5, n_dwumianowy + 1.5, 1), edgecolor='black', alpha=0.7)
plt.title('Rozkład Dwumianowy')

plt.subplot(1, 3, 3)
plt.hist(proby_poisson, bins=np.arange(-0.5, max(proby_poisson) + 1.5, 1), edgecolor='black', alpha=0.7)
plt.title('Rozkład Poissona')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import describe
```

```python
stats_bernoulli = describe(proby_bernoulli)
stats_dwumianowy = describe(proby_dwumianowy)
stats_poisson = describe(proby_poisson)
```

```python
print("Statystyki dla Rozkładu Bernoulliego:")
print(f"Średnia: {stats_bernoulli.mean}")
print(f"Wariancja: {stats_bernoulli.variance}")
print(f"Kurtoza: {stats_bernoulli.kurtosis}")
print(f"Skośność: {stats_bernoulli.skewness}")
```

```python
print("Statystyki dla Rozkładu Dwumianowego:")
print(f"Średnia: {stats_dwumianowy.mean}")
print(f"Wariancja: {stats_dwumianowy.variance}")
print(f"Kurtoza: {stats_dwumianowy.kurtosis}")
print(f"Skośność: {stats_dwumianowy.skewness}")
```

```python
print("Statystyki dla Rozkładu Poissona:")
print(f"Średnia: {stats_poisson.mean}")
print(f"Wariancja: {stats_poisson.variance}")
print(f"Kurtoza: {stats_poisson.kurtosis}")
print(f"Skośność: {stats_poisson.skewness}")
```

```python
from scipy.stats import binom
```

```python
n_dwumianowy = 20
p_dwumianowy = 0.4
values_dwumianowy = np.arange(0, n_dwumianowy + 1, 1)
```

```python
pmf_dwumianowy = binom.pmf(values_dwumianowy, n_dwumianowy, p_dwumianowy)
pmf_dwumianowy
```

```python
plt.bar(values_dwumianowy, pmf_dwumianowy, color='orange', alpha=0.7)
plt.title('Rozkład Dwumianowy')
plt.xlabel('Wartość (k)')
plt.ylabel('Prawdopodobieństwo')
```

```python
suma_prawdopodobienstw = np.sum(pmf_dwumianowy)
print(f"Suma prawdopodobieństw: {suma_prawdopodobienstw}")
```

```python
np.random.seed(42)

srednia = 0
odchylenie_standardowe = 2
liczba_danych = 100
```

```python
dane_normalne = np.random.normal(srednia, odchylenie_standardowe, liczba_danych)
dane_normalne
```

```python
statystyki_opisowe = describe(dane_normalne)
```

```python
print("Statystyki opisowe dla rozkładu normalnego:")
print(f"Średnia: {statystyki_opisowe.mean}")
print(f"Wariancja: {statystyki_opisowe.variance}")
print(f"Kurtoza: {statystyki_opisowe.kurtosis}")
print(f"Skośność: {statystyki_opisowe.skewness}")
```

```python
dane_normalne_10000 = np.random.normal(srednia, odchylenie_standardowe, 10000)
dane_normalne = [dane_normalne_10000[:100], dane_normalne_10000[:1000], dane_normalne_10000 ]
statystyki = [ describe(x) for x in dane_normalne]
for i in range(len(statystyki)):
    print(f"\nStatystyki opisowe dla 100{'0'*i} danych:")
    print(f"Średnia: {statystyki[i].mean}")
    print(f"Wariancja: {statystyki[i].variance}")
    print(f"Kurtoza: {statystyki[i].kurtosis}")
    print(f"Skośność: {statystyki[i].skewness}")
```

```python
from scipy.stats import norm
```

```python
np.random.seed(42)

mu_1, sigma_1 = 1, 2
mu_2, sigma_2 = -1, 0.5
```

```python
dane_normalne = np.random.normal(mu_1, sigma_1, 1000)
```

```python
plt.figure(figsize=(15,5))

plt.hist(dane_normalne, bins=30, density=True, alpha=0.7, color='blue', label='Rozkład normalny dla μ=1, σ=2')

x = np.linspace(mu_1 - 4 * sigma_1, mu_1 + 4 * sigma_1, 100)
plt.plot(x, norm.pdf(x, mu_1, sigma_1), 'r-', lw=2, label='Standardowy rozkład normalny')

x_new = np.linspace(mu_2 - 4 * sigma_2, mu_2 + 4 * sigma_2, 100)
plt.plot(x_new, norm.pdf(x_new, mu_2, sigma_2), 'g-', lw=2, label='Gęstość prawdopodobieństwa dla μ=-1, σ=0.5')

plt.legend(loc='upper right')

plt.title('Histogram i Wykresy Gęstości Rozkładu Normalnego')
plt.xlabel('Wartość')
plt.ylabel('Gęstość prawdopodobieństwa')
```
